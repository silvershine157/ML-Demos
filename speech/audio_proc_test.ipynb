```python
import torchaudio
import IPython.display as ipd
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt

from dataset import *
```

```python
def test1():
    sr = 22050
    ds = LinSpecDataset(torchaudio.datasets.LJSPEECH('./data'), limit=1)
    orig_wav, _, _, _ = ds.base_ds[0]
    orig_S = ds.wav_to_spec(orig_wav)
    print(orig_S.shape)
    
    print(orig_S.min(), orig_S.max())
    librosa.display.specshow(orig_S.numpy().T)
    plt.show()
    recon_wav = ds.spec_to_wav(orig_S)
    librosa.output.write_wav('data/audio_sample.wav', recon_wav.squeeze().numpy(), sr)
test1()
ipd.Audio('data/audio_sample.wav')
```
